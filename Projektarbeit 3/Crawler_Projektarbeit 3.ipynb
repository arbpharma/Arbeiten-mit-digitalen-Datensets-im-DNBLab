```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
import requests
```

```python
url = 'https://portal.dnb.de/opac/showFirstResultSite?currentResultId=cod%3Dnt%26any%26onlinefree%26scores&currentPosition=10'
download_path = "D:/Projektarbeit3_Files/"
```

```python
chrome_options = Options()
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument("--headless")
chrome_options.add_experimental_option("prefs", {
    "download.default_directory": download_path,  
    "download.prompt_for_download": False,        
    "download.directory_upgrade": True,           
    "safebrowsing.enabled": True                  
})

driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
driver.get(url)
element = driver.find_element(By.ID, "recordLink_0")
element.click()
time.sleep(2)

# falls man alle daten downloaden will
# number_of_hits = int(driver.find_element(By.CLASS_NAME, "amount").text.split(" ")[-1]) 
# wir downloaden nur die ersten 100, weil das allein schon 5 gb groß ist. ganzes datenset ~174GB, 5 stunden download!
number_of_hits = 100 
for i in range(0, number_of_hits):
    try:
        pdf_link = driver.find_element(By.ID, "linkOnlineNetpub").find_element(By.TAG_NAME, "a").get_attribute("href")
        time.sleep(1)
        response = requests.get(pdf_link)
        if response.status_code == 200:
            with open(download_path + f"file_{i}.pdf", 'wb') as file:
                file.write(response.content)
    except:
        print(f"error at file {i}")
    driver.find_element(By.XPATH, "//*[@title='zum nächsten Treffer blättern']").click()
    time.sleep(1)
    print(f"downloaded file {i} of {number_of_hits}!")
```
