```python
import fitz  # PyMuPDF
from PIL import Image
import pytesseract
import io

# Path to Tesseract executable (only needed if it's not in your PATH)
# Update the path below for your system if necessary
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def extract_first_image_and_text(pdf_path, output_image_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    
    # Access the first page
    first_page = pdf_document[0]
    
    # Extract images from the first page
    images = first_page.get_images(full=True)
    if not images:
        print("No images found on the first page.")
        return None, None
    
    # Get the first image
    first_image_ref = images[0][0]
    base_image = pdf_document.extract_image(first_image_ref)
    
    # Convert the image data to a Pillow Image object
    image_bytes = base_image["image"]
    image = Image.open(io.BytesIO(image_bytes))
    
    # Save the image to a file
    image.save(output_image_path)
    print(f"Image saved to {output_image_path}")
    
    # Perform OCR to extract text from the image
    text = pytesseract.image_to_string(image)
    return text #,image

# Define the PDF path and output image path
pdf_path = "pdf/2254.pdf"

output_image_path = "output_image.jpg"

# Call the function
extracted_text = extract_first_image_and_text(pdf_path, output_image_path)

# Display the results
if image:
    print("Extracted Text:")
    print(extracted_text)
```
