```python
import cv2
import fitz  # PyMuPDF
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Users\Megaport\AppData\Local\Programs\Tesseract-OCR\tesseract.exe"
from matplotlib import pyplot as plt
import os

# Schritt 1: PDF in Bilder umwandeln
def pdf_to_images(pdf_path, output_folder="output_images"):
    os.makedirs(output_folder, exist_ok=True)
    pdf_document = fitz.open(pdf_path)
    for page_number in range(len(pdf_document)):
        page = pdf_document.load_page(page_number)
        pix = page.get_pixmap()
        image_path = os.path.join(output_folder, f"page_{page_number}.png")
        pix.save(image_path)
    return output_folder

# Schritt 2: Erkenne Notenlinien und Noten
def process_music_sheet(image_path):
    # Bild laden
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    _, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)

    # Notenlinien finden
    horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (40, 1))
    detected_lines = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

    # Ziehe die Linien ab, um die Noten zu isolieren
    notes_only = cv2.subtract(thresh, detected_lines)

    # Konturen finden (mögliche Noten)
    contours, _ = cv2.findContours(notes_only, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    results = []
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        if 10 < w < 50 and 10 < h < 50:  # Filter für Notengröße
            note_roi = img[y:y + h, x:x + w]
            # OCR auf das Noten-Bild anwenden
            text = pytesseract.image_to_string(note_roi, config='--psm 10')
            results.append((x, y, w, h, text.strip()))
            # Zeichne die gefundenen Noten (Debugging)
            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Ergebnis anzeigen
    plt.figure(figsize=(10, 10))
    plt.imshow(img, cmap='gray')
    plt.title("Gefundene Noten")
    plt.show()

    return results

# Schritt 3: Hauptprogramm
pdf_path = "pdfs/missa-phrygia.pdf"
output_folder = pdf_to_images(pdf_path)

for image_file in os.listdir(output_folder):
    if image_file.endswith(".png"):
        image_path = os.path.join(output_folder, image_file)
        print(f"Analysiere {image_file}...")
        notes = process_music_sheet(image_path)
        print(f"Gefundene Noten in {image_file}: {notes}")
```
