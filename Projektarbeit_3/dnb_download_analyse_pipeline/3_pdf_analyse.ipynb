```python
import nltk
nltk.download('stopwords')
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
german_stop_words = stopwords.words('german')
import fitz 
import os
```

```python
pdf_path = "pdf/"
dirr = os.listdir(pdf_path)
def extract_text_from_pdf(pdf_path):
    try:
        pdf_document = fitz.open(pdf_path)
        if pdf_document.page_count > 0:
            first_page = pdf_document[0]
            text = first_page.get_text()
            if text.strip():
                return text
    except:
        print(pdf_path.split("/")[-1] + " not readable")
```

```python
# extract text from pdf + print the output
```

```python
# this version is without OCR... you can get the OCR Function from 2_pdf_reader_base_code.ipynb
stringg = ""
for file in dirr:
    print("--------------------------")
    print()
    print()
    print()
    ret = extract_text_from_pdf(pdf_path + file)
    if ret != None and type(ret) == type(""):
        ret = ret.split(" ")
        rett = ""
        for word in ret:
            if word.lower() not in german_stop_words:
                rett += word.lower() + " "
        
        rett = rett[:-1]
        stringg += rett.replace("\n", " ").replace("œ", " ")
    print(ret)
    print()
    print()
    print()
    print("--------------------------")
```

```python
# UNCLEANED TEXT
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(stringg)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
